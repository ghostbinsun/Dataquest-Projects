#### I want to aggregate on volkswagen, bmw, mercedes_benz, opel, and audi

```python
autos['brand'].value_counts().head()
```

```python
selected_brand = ['volkswagen', 'bmw', 'mercedes_benz', 'opel', 'audi']
```

```python
brand_mean_P = {}
for i in selected_brand:
    brand_mean_P[i] = round(autos.loc[autos['brand'] == i, 'price'].mean(), 1)
# round(, 1), round to 2 decimal places
```

```python
brand_mean_P
```

#### ---Created a dictionary to store mean price for top 5 popular brands

#### Observation: audi is the most expensive brand, although it has the least amount of cars

```python
bmp_series = pd.Series(brand_mean_P)
df = pd.DataFrame(bmp_series, columns = ['mean_price'])
df
# change dictionary to dataframe:
#    1. if you want the keys to be index, then, change them to series first, then they will look like vertical with keys as the index
#    2. if you want the keys to be columns, then, change them directedly from (dictionary), e.g. two lines below
```
